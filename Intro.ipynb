## While Loop (less Pythonic)

```python
s = "RacecAr "

def is_palindrome1(string):
    clean_string = string.lower().replace(' ', '')
    first, last = 0, len(clean_string)-1
    while( first < last):
        if clean_string[first] == clean_string[last]:
            first +=1
            last -=1
        else:
            return False
    
    return True

is_palindrome1(s) 
```

## String Slicing (More Pythonic)

```python
def is_palindrome2(string):
    clean_string = string.lower().replace(' ', '')
    reversed_string = clean_string[::-1]
    return clean_string == reversed_string

s = " Rac eC ar "
is_palindrome2(s)
```

```python
import csv


directors = {}


with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        current_counter = directors.get( row['director_name'], 0)
        if current_counter == 0:
            directors[row['director_name']] = 1
        else:
            directors[row['director_name']] += 1

    for director in directors:
        print(f"{director} has made {directors.get(director)} movie(s).")
        print("")

```

```python
import csv

movies = {}
sum = 0


with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        genre = row['genres']
        if genre[0:6] == ("Action"):
            movies[row['movie_title']] = row['duration'] 
    movie_durations = movies.values()

    
    for duration in movie_durations:
        if duration != '':
            new_sum = int(duration) + sum
            sum = new_sum
    
    the_average = int(sum / len(movie_durations))
    print(f"The average duration of action movies is about {the_average} minutes")
```
