## 1. How many movies have "Star Wars in the movie_title field?

```python
import csv

Star_Wars = []

with open("movie.csv", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # goes through movie title row in csv 
        movie_title = row['movie_title']
        if movie_title != '':
            #if the first 8 letters are star wars
            if movie_title[0:9] == ("Star Wars"):
                #if the movie has star wars in the name, added to list Star_Wars
                Star_Wars.append(movie_title)
                
    print(f"There are {len(Star_Wars)} Star Wars movies")
                
```

## 2. Which movie has the highest IMDB Score?

```python
import csv

ratings = {}
dic_movies = []
dic_ratings = []
real_dic_ratings = []

with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    #goes through movie title row and gives each the value of its associated imdb score
    for row in reader:
        ratings[row['movie_title']] = float(row['imdb_score'])

    # all movies added to a list to get movie index
    for movie in ratings:
            dic_movies.append(movie)
    
    #list of imdb values 
    dic_ratings = ratings.values()

    
    #for each score value, it is added to  searched list 
    for value in dic_ratings:
           real_dic_ratings.append(value)
    
    #find highest score of that list
    ratings_index = real_dic_ratings.index(max(ratings.values()))

    #used for printed highest scored movie 
    highest_movie = dic_movies[ratings_index]
    highest_rating = max(ratings.values())

    print(f"{highest_movie} has the highest IMDB score of  {highest_rating}")
```

## 4. For each director, how many movies have they directed (from this dataset)?

```python
import csv


directors = {}


with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #go through row director name and gives it value of 0 
        current_counter = directors.get( row['director_name'], 0)

        #if the key has not appeared, value is 1
        if current_counter == 0:
            directors[row['director_name']] = 1
        
        #if the key has appeared, added to the number of times it has appeared
        else:
            directors[row['director_name']] += 1

    for director in directors:
        print(f"{director} has made {directors.get(director)} movie(s).")
        print("")
```

## 5. What's the average duration of Action Movies?

```python
import csv

movies = {}
sum = 0


with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:

        #goes through the row of genres
        genre = row['genres']

        #if the word Action appears in this category, it is added to the dict of movies
        if genre[0:6] == ("Action"):
            movies[row['movie_title']] = row['duration'] 

    #list of each action movie duration
    movie_durations = movies.values()

    #the sum of all durations of action movies, if they exist
    for duration in movie_durations:
        if duration != '':
            new_sum = int(duration) + sum
            sum = new_sum
    
    #the average of action movie durations
    the_average = int(sum / len(movie_durations))
    print(f"The average duration of action movies is about {the_average} minutes")
```
